17 in-built datasets

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset('tips')
```

```python
data.head(7)
```

Plot a Scatter Plot between total_bill and tip

```python
sns.relplot(x="total_bill",y="tip",data=data,kind="scatter")
```

Amount of bill = greater tip. 
Therefore, linear relationship

hue parameter - shows a column which has specific 2 or more values

```python
sns.relplot(x="total_bill",y="tip",data=data,kind="scatter",hue="smoker")
```

style parameter - similar as hue, overplots on hue parameter with other column value

size parameter

```python
sns.relplot(x="total_bill",y="tip",data=data,hue = "smoker",style="sex",size="time")
```

```python
sns.relplot(x="total_bill",y="tip",data=data,hue = "smoker",style="sex",size="time")
```

```python
sns.relplot(x="total_bill",y="tip",data=data,hue = "smoker",style="sex",size="time",sizes=(10,50))
```

kind = line

```python
sns.relplot(x="total_bill",y="tip",data=data,kind="line",hue = "smoker",style="sex",size="time")
```

Violin Plot

- combination on boxplot and distribution of data
- box : Interquatile range
- circle : median
- distribution up and down are same like mirroring
- lines : whiskers
- start point of line : minimum
- end end of line : maximum
- After max point : outliers
- Box start : 25%
- Box middle : 50%
- Box end : 75%

```python
plt.style.use("fivethirtyeight")
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.violinplot(tips["tip"])
```

```python
sns.catplot(x="day",y="tip",data=tips,kind="violin")
```

```python
sns.catplot(x="day",y="tip",data=tips,kind="violin",hue="sex")
```

```python
#split = True //both side of distribution different columns of hue 
sns.catplot(x="day",y="tip",data=tips,kind="violin",hue="sex",split=True)
```

barplot and countplot
- numerical data wrt categorical data

Barplot
- black line: error bar
  * more bigger error bar, less reliable

```python
sns.catplot(x="day",y="tip",data=tips,kind="bar")
```

```python
sns.catplot(x="day",y="tip",data=tips,kind="bar",hue="sex")
```

countplot
- takes only 1 column

```python
sns.catplot(x="day",data=tips,kind="count",hue="sex")
```

```python
sns.barplot(tips["tip"]) #univariant
```

HeatMaps

```python
flights = sns.load_dataset("flights")
```

```python
flights.head()
```

```python
x = flights.pivot_table(index="year",columns="month",values="passengers",aggfunc="sum")
```

```python
x
```

```python
#darker the color, less passengers
sns.heatmap(x)
```

Cluster Map

```python
#combines 2 similar column or row
sns.clustermap(x)
```
