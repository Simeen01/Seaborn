```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use("fivethirtyeight")
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head(10)
```

```python
titanic.tail(10)
```

```python
titanic.shape
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
#each column's sum of null values
titanic.isnull().sum()
```

```python
#Total no. of null values
titanic.isnull().sum().sum()
```

```python
titanic.fillna(method="bfill",inplace=False)
```

```python
titanic.interpolate(method="linear")
```

Survival class

```python
survivals_per_class = titanic[["pclass","survived"]].groupby(["pclass"],as_index=False).mean().sort_values(by="survived",ascending=False)
```

```python
survivals_per_class
```

```python
survivals_per_age = titanic[["age","survived"]].groupby(["age"],as_index=False).sum().sort_values(by="survived",ascending=False)
survivals_per_age
```

Survival gender

```python
survivals_per_sex = titanic[["sex","survived"]].groupby(["sex"],as_index=False).mean().sort_values(by="survived",ascending=False)
survivals_per_sex
```

Survived as per sibsp

```python
survivals_per_sibsp = titanic[["sibsp","survived"]].groupby(["sibsp"],as_index=False).mean().sort_values(by="survived",ascending=False)
survivals_per_sibsp
```

Survival per parch

```python
survived_per_parch = titanic[["parch","survived"]].groupby(["parch"],as_index=False).mean().sort_values(by="survived",ascending=False)
survived_per_parch
```

```python
#plt.subplot(1,3,1)
sns.barplot(y="age",x="survived",data=titanic)
```

Lesser people survived

```python
sns.barplot(y="age",x="survived",data=titanic,hue="sex")
```

```python
sns.barplot(y="age",x="survived",data=titanic,hue="pclass")
```

```python
sns.catplot(x="survived",data=titanic,kind="count",hue="pclass")
```

pclass 3 had most passengers but most did not survive.

pclass 1 had the highest survival rate

```python
sns.lineplot(x="pclass",y="survived",data=titanic,hue="embarked", style="sex")
```

```python
sns.catplot(x="survived",y="fare",data=titanic,kind="bar")
```

Passengers with higher fare had higher rate of survival.
